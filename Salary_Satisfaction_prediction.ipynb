#The goal is to predict whether the salary for data-related positions, based on location, company reviews, and job title, is satisfactory or not

#Import libraries

```python
from sklearn.model_selection import train_test_split
import pickle
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

#Import dataset

```python
from google.colab import files
uploaded = files.upload()
```

#Read the data

```python
import pandas as pd
df = pd.read_csv('preprocessed_df.csv')
```

```python
df
```

#Feature Engineering

```python
df['Average Salary'].median()
```

```python
# This code classifies salaries into two categories based on a threshold (102).
# If the 'Average Salary' is 102 or more, it is labeled as '1' (satisfied); otherwise, it is labeled as '0' (not satisfied).

df['satisfaction'] = df['Average Salary'].apply(lambda x: 1 if x >= 102 else 0)
```

```python
df[['satisfaction', 'Average Salary']]
```

```python
x = df.drop(columns = ['satisfaction', 'Average Salary'])
```

```python
y = df['satisfaction']
```

#Modeling

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state = 101)
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(x_train, y_train)
```

```python
prediction = logmodel.predict(x_test)
```

```python
prediction
```

```python
confusion_matrix(y_test, prediction)
```

```python
len(y_test)
```

```python
print(classification_report(y_test, prediction))
```

```python
accuracy_score(y_test, prediction)
```

#Save the model

```python
pickle.dump(logmodel, open('SalarySatisfaction_Estimator.pkl' , 'wb'))
```
