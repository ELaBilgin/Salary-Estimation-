#Data cleaning

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.replace('$', ' '))
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.replace('K', ' '))
```

```python
df = df[df['Salary Estimate']!= '-1']
```

```python
df = df[~df['Salary Estimate'].str.contains('Per Hour')]
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split())
```

```python
df['Min Salary'] = df['Salary Estimate'].apply(lambda x: x[0])
df['Max Salary'] = df['Salary Estimate'].apply(lambda x: x[2])
```

```python
df['Min Salary'] = df['Min Salary'].astype(int)
df['Max Salary'] = df['Max Salary'].astype(int)
df['Average Salary'] = (df['Min Salary'] + df['Max Salary']) / 2
```

```python
df['State'] = df['Location'].apply(lambda x: x.split()[-1])
```

```python
df = df[df['State'] != 'Kingdom']
```

```python
df = df[df['Rating']!= -1]
```

```python
def title_organizer(x):
  x = x.lower()
  if ('data scientist' in x) or ('data science' in x):
    return 'data scientist'
  elif 'data engineer' in x:
    return 'data engineer'
  elif 'deep learning' in x:
    return 'deep learning engineer'
  elif 'machine learning'in x:
    return 'machine learning engineer'
  elif 'director' in x:
    return 'director'
  elif 'manager' in x:
    return 'manager'
  elif 'analyst' in x:
    return 'analyst'
  elif 'researcher' in x:
    return 'researcher'
  else:
    return 'other'
```

```python
df['Job Title'] = df ['Job Title'].apply(title_organizer)
```

```python
df.drop(columns = ['Salary Estimate','Company Name','Location','Size', 'Founded', 'Type of ownership','Industry','Sector', 'Revenue', 'Min Salary', 'Max Salary'], inplace = True)
```
